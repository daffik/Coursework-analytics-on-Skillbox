```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

df_progress_phases = pd.read_csv("progress_phases.csv")
df_progresses = pd.read_csv("progresses.csv")
df_progresses.rename(columns = {'id' : 'progress_id'}, inplace = True)
df_courses = pd.read_csv("courses.csv", index_col=0)
df_courses.rename(columns = {'id' : 'course_id', 'title' : 'course_title', 'field' : 'course_field'}, inplace = True)
df_course_contents = pd.read_csv("course_contents.csv")

df_students = pd.read_csv("students.csv")

df = df_progress_phases
df = df.merge(df_progresses, how='left', on='progress_id')
df = df.merge(df_courses, how='left', on='course_id')
df = df.merge(df_course_contents, how='left', on=['course_id', 'module_number', 'lesson_number'])
df.head()
#df[df.student_id == '768c2987a744c51ce64a5993a2a94eaf']
```

```python
#общее кол-во курсов

df['course_id'].nunique()
```

```python
#количество модулей на каждом курсе

df.groupby('course_title')['module_number'].agg(module_count = 'max').reset_index().sort_values('module_count')
```

```python
#количество уроков в каждом модуле на каждом курсе

df_lesson = df.groupby(['course_title', 'module_title'])['lesson_title'].nunique().reset_index()
df_lesson
```

```python
#медианное количество уроков в модуле на каждом курсе

df_lesson.groupby('course_title')['lesson_title'].median().reset_index()
```

```python
# количество учеников на каждом курсе

df_stud_on_cour = df.groupby('course_title')['student_id'].agg(student_count='nunique').reset_index()
df_stud_on_cour.sort_values('student_count')
```

```python
# минимальный, максимальный, средний, медианный возраст студентов
df_students['birthday'] = pd.to_datetime(df_students['birthday'], errors="coerce")
df_students['age'] = (pd.to_datetime('today') - df_students['birthday']).astype("<m8[Y]")
df_students = df_students[(df_students['age'] > 6) & (df_students['age'] < 65)]
print(f'Минимальный возраст: {int(df_students.age.min())} \
      \nМаксимальный возраст: {int(df_students.age.max())} \
      \nСредний возраст: {int(df_students.age.mean())} \
      \nМедианный возраст: {int(df_students.age.median())}')
```

```python
# минимальный, максимальный, средний, медианный возраст студентов на каждом курсе
df_students.rename(columns = {'id' : 'student_id'}, inplace = True)
df1 = df.merge(df_students, how='left', on='student_id')
df1.groupby('course_title')['age'].agg(max_age = 'max', min_age = 'min', mean_age = 'mean', median_age = 'median')
```

```python
# Постройте bar-chart, отражающий количество студентов на каждом курсе. 
# Ticks нужно развернуть так, чтобы они были читаемы
fig, ax = plt.subplots(figsize=(18, 4))

ax.bar(df_stud_on_cour['course_title'], df_stud_on_cour['student_count'])
ax.set_title("Кол-во студентов на каждом курсе")
ax.set_ylabel('человек')
ax.set_xlabel('Название курса')
plt.setp(ax.get_xticklabels(), rotation=45, ha='right', size=10)
plt.show()
```

```python
# (m) Постройте горизонтальный (столбцы должны располагаться горизонтально) bar-chart, 
# отражающий количество студентов на каждом курсе. График должен иметь заголовок. 
# Значения должны быть отсортированы. Цвет столбцов должен содержать информацию о сфере, 
# к которой относится курс (то есть нужна легенда). Прозрачность должна стоять на отметке 0.1. 
# На график должна быть нанесена линия медианы. У медианы должен быть свой цвет. Рамки у графика быть не должно ⭐


data = df.groupby(['course_title', 'course_field'])['student_id'].agg(student_count='nunique')\
    .reset_index()\
    .sort_values('student_count',ascending=False)

fig, ax = plt.subplots(figsize=(10, 5))
sns.barplot(data=data, y='course_title', x='student_count', hue='course_field', dodge=False, alpha=0.1)
ax.axvline(x=data['student_count'].median(), color='red', ls=':')
ax.set_title("Кол-во студентов на каждом курсе")
ax.set_xlabel('человек')
ax.set_ylabel('Название курса')
plt.box(on=None)
plt.legend(loc='best')
plt.show()
```

```python
# Рассчитать прирост студентов на каждом курсе в каждом месяце за всю историю 
# (каждый месяц в диапазоне от марта 2016 до июля 2019 включительно). 
# Считать дату начала прохождения курса студентом по дате начала первой домашней работы.

#группируем по названию курсу, номеру модуля, ищем первый урок с домашней работой,
# еще раз групируем по названию курса и выбираем первую строчку

df3 = df[(df['is_homework'] == True)]\
    .groupby(['course_title', 'module_number'])['lesson_number'].agg('min').reset_index()\
    .groupby('course_title').first().reset_index()

# к получившейся таблице добавляем остальные данные
df_growth = df3.merge(df, how='left', on=['course_title', 'module_number', 'lesson_number'])

#столбец с датой переформатируем в год-месяц, группируем по дате и названию курса, считаем кол-во студентов
df_growth['start_date'] =  pd.to_datetime(df_growth['start_date']).dt.strftime('%Y-%m')

# df_growth['start_date'] = df_growth['start_date'].astype('datetime64[M]').dt.strftime('%Y-%m')
df_growth = df_growth.groupby(['start_date', 'course_title'])['student_id'].count().reset_index()
df_growth = df_growth.sort_values(['course_title', 'start_date'])

#добавляем столбец с приростом студентов по месяцам по курсам
# for item in df_growth['course_title'].unique():
#     indexes = df_growth.loc[df_growth['course_title']==item, 'student_id'].keys()
#     for index in indexes:
#         df_growth.loc[index, 'student_growth'] = df_growth.loc[indexes[0]:index, 'student_id'].sum()

df_growth
```

```python
# Рассчитать минимальное, максимальное, среднее, медианное время прохождения каждого модуля
# (разность между временем начала и окончания выполнения домашней работы) для каждого курса. 
# Если домашних заданий в модуле несколько, то считать разность между временем начала выполнения первой домашней работы
# и временем окончания выполнения последней домашней работы в модуле

#фильтруем датасет по домашним работам
df_problems = df[df['is_homework'] == True]

#убираем строки с пропущенными значениями
df_problems = df_problems.dropna()

#выбираем минимальную дату старта начала дз в модуле и максимальную дату сдачи дз в модуле
df_problems = df_problems.groupby(['course_title', 
                                   'module_title',
                                   'module_number',
                                   'student_id']).agg({'start_date': 'min', 
                                                       'finish_date': 'max'}).reset_index()

#преобразовываем даты в формат дататайм
df_problems['finish_date'] =  pd.to_datetime(df_problems['finish_date'])
df_problems['start_date'] =  pd.to_datetime(df_problems['start_date'])

#находим период работы над дз в каждом модуле и округляем до дней
df_problems['module_time'] = (df_problems['finish_date'] - df_problems['start_date']).dt.days

#убираем из датасета некорректные данные, где дата старта больше даты окончания работы над дз
df_problems = df_problems[df_problems['finish_date'] > df_problems['start_date']]

#рассчитываем минимальное, максимальное, среднее, медианное время работы над дз в каждом модуле для каждого курса.
df_problems_info = df_problems.groupby(['course_title',
                                        'module_number',
                                        'module_title'])['module_time'].agg(min_time='min', 
                                                                            max_time='max', 
                                                                            mean_time='mean',  
                                                                            median_time='median').reset_index()
df_problems_info.head()
```

```python
# На основании первого пункта построить line-graph с медианным временем прохождения каждого модуля для каждого курса.
# 15 графиков. Графики должны иметь заголовки

for course in df_problems_info.course_title.unique():
    data = df_problems_info[df_problems_info['course_title'] == course]
    fig, ax = plt.subplots(figsize=(10,5))
    sns.lineplot(data=data, x='module_title', y='median_time', ax=ax, ci=None)
    ax.set_title(f'Медианное время прохождение модулей на курсе: {course}')
    ax.set_ylabel('Время, дни')
    ax.set_xlabel(None)
    plt.xticks(rotation=90, ha='right')

plt.show()
```

```python
# Чтобы выявить сезонность, посчитать медианное время выполнения домашней работы по месяцам 
# (12 месяцев, январь-декабрь) для каждого курса. 

#добавляем в датасет месяц сдачи дз
df_problems['month_end'] = df_problems['finish_date'].dt.month

# создаем датасета с медианым временем выполнения домашней работы по месяцам, по дате сдачи дз
df_seasonality_end = df_problems.groupby(['course_title',
                                            'month_end'])['module_time'].agg(min_time='min', 
                                                                               max_time='max', 
                                                                               mean_time='mean',  
                                                                               median_time='median').reset_index()
```

```python
# На основании третьего пункта построить line-graph, 
# на который будут нанесены линии для каждого курса с медианным временем выполнения домашней работы по месяцам. 
# 15 линий на графике. График должен иметь заголовок. Ось, отражающая время прохождения, должна быть подписана. 
# Линия для каждого курса должна иметь свой цвет (нужна легенда). Рамок у графика быть не должно ⭐

df_seasonality_end
fig, ax = plt.subplots(figsize=(10,5))
sns.lineplot(data=df_seasonality_end, x='month_end', y='median_time', hue='course_title')
ax.set_title('Медианное время выполнения домашней работы по месяцам по курсам')
ax.set_ylabel('Медианное время прохождения, дни')
ax.set_xlabel('мес.')
ax.set_xticks(range(1, len(df_seasonality_end['month_end'].unique())+1))

plt.grid(color='grey',
         linestyle='-',
         linewidth=1,
         alpha = 0.2)
plt.box(on=None)
plt.legend(bbox_to_anchor=(1, 1), fontsize=13, frameon=False)
plt.show()
```

## 4. Расчет конверсии

2. **(m)** Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы

Прозрачность должна стоять на отметке 0.1. На графики должна быть нанесена линия медианы конверсии для каждого курса. У медианы должен быть свой цвет. Рамок у графиков быть не должно ⭐
4. На основании рассчитанных значений сделайте аналитический вывод (должен быть полным и покрывать все полученные выше метрики)

```python
#создаем датасет со студенами, которые закончили дз
df_finish = df[(df['is_homework'] == True) & (df['status'] == 'done')]
df_finish = df_finish[['course_title',
                       'course_field',
                       'module_number',
                       'module_title', 
                       'student_id']]

#считаем количество студентов, которые закончили дз, на каждом курсе на каждом модуле
df_finish = df_finish.groupby(['course_title', 
                               'module_title', 
                               'module_number'])['student_id'].count().reset_index()
#переименовываем стобец для удобства
df_finish.rename(columns = {'student_id' : 'students_finish'}, inplace=True)

#сортируем датасет по курсу и по номеру модуля
df_finish.sort_values(['course_title', 'module_number'], inplace=True)

#создаем датасет со всеми студентами, которые начинали делать дз
df_start = df[df['is_homework'] == True]
df_start = df_start[['course_title',
             'course_field',
             'module_number',
             'module_title',
             'student_id']]

#считаем количество студентов, которые начинали дз, на каждом курсе на каждом модуле
df_start = df_start.groupby(['course_title',
                             'course_field',
                             'module_title',
                             'module_number'])['student_id'].count().reset_index()

#переименовываем стобец для удобства
df_start.rename(columns = {'student_id' : 'students_start'}, inplace=True)

#сортируем датасет по курсу и по номеру модуля
df_start.sort_values(['course_title', 'module_number'], inplace=True)

#соединяем 2 датасета
df_conversion = df_start.merge(df_finish, how='left', on=['course_title', 'module_title', 'module_number'])

#сдвигаем столбец, с количеством студентов сдавших дз, вниз на одну строчку
df_conversion['students_finish'] = df_conversion['students_finish'].shift(1)

#удаляем из датасета все первые модули
df_conversion = df_conversion[df_conversion['module_number'] != 1]

#создаем столбец с конверсией
df_conversion['conversion'] = (df_conversion['students_start'] / df_conversion['students_finish']) * 100
df_conversion.head(10)
```

```python
# Постройте bar-chart, отражающий конверсию перехода студентов из одного модуля в другой на каждом курсе. 
# График должен иметь заголовок. Ticks нужно развернуть так, чтобы они были читаемы

df_conversion_bar = df_conversion.groupby('course_title')['conversion'].mean().reset_index()
df_conversion_bar

fig, ax = plt.subplots(figsize=(15, 5))
ax.bar(df_conversion_bar['course_title'], df_conversion_bar['conversion'])
ax.set_title("Средняя конверсия перехода студентов из одного модуля в другой на каждом курсе")
ax.set_ylabel('конверсия, %')
plt.setp(ax.get_xticklabels(), rotation=45, ha='right', size=10)
plt.show()
```

&nbsp;&nbsp;&nbsp;&nbsp;*(необязательное задание)*

Иногда студенты берут курсы, которые оказываются для них неподъемными. Это может быть как по причинам недостаточной изначальной подготовки, так и по причинам, связанным с низкой мотивацией студента. Для улучшения качества контента полезно выявить причину. На основании имеющихся данных придумайте метрику успеваемости студента. **Обоснуйте ее.** Выявите таких студентов на каждом курсе, чтобы предоставить продюсерам список проблемных студентов. 

```python
#сортируем по статусу выполнения дз status = fail
df_fail = df[df['status'] == 'fail']
#оставляем нужные столбцы
df_fail = df_fail[['student_id', 'course_title', 'module_title', 'module_number', 'start_date']]
#преобразуем дату старта в формат дат
df_fail['start_date'] = pd.to_datetime(df_fail['start_date'])
#вычисляем сколько дней уже не выполнена дз и округляем
df_fail['time_for_hw'] = (pd.to_datetime(df['start_date'].max()) - df_fail['start_date']).dt.days
#группируем таблицу по студентам, время невыполнения дз суммируем, сортируем по убыванию
df_fail = df_fail.groupby('student_id')['time_for_hw'].sum().reset_index().sort_values('time_for_hw', ascending=False)
#фильтруем, оставляем только тех студентов, у кого суммарное кол-во дней невыполнения всех дз больше медианного значения
df_fail = df_fail[df_fail['time_for_hw'] > df_fail['time_for_hw'].median()]
#создаем новый датасет, фильтруем по новому стобцу
df_problems_stud = df.merge(df_fail, how='left', on='student_id')
df_problems_stud = df_problems_stud[df_problems_stud['time_for_hw'] > 1]
#группируем по курсам и получаем список студентов для каждого курса
df_problems_stud.groupby('course_title')['student_id'].unique().reset_index()
```

Мы выявили студентов, которые потратили наибольшее количество времени на несданные дз, по каждому курсу
